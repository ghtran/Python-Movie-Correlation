```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Configurations of the plots we will create

# Read in the data
# df = data frame
df = pd.read_csv(r'D:\Documents\Data Analysis\Personal Projects\Movie Data Project\movies.csv')
```

```python
# Exploring data
df
```

```python
## Data Cleaning
```

```python
# See if there's missing data

# using a for loop to look for percent of missing information in each column
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Check the data types in each columns

print(df.dtypes)
```

```python
# Checking for outliers

df.boxplot(column = ['gross'])
```

```python
# Drop Duplicates
df.drop_duplicates()
```

```python
## Gross Value Correlations
```

```python
# Organizing data so that the gross is ordered in descending order

df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')

# sns.regplot(x = 'gross', y = 'budget', data = df)
```

```python
df.head()
```

```python
# Making a regression to see if there is any correlation between budget vs gross 

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"red"})
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
```

```python
## Correlation Table and Heat Map
```

```python
# Look at corrlation, Pearson(default), spearman, kendall

df.corr(method='pearson')
```

```python
# Correlation using Spearman method 
df.corr(method='spearman')
```

```python
# Making a heat map/correaltion matrix
corr_matrix=df.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")

plt.show()
```

```python
# Assinging a numeric value for each unique column category
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
# 
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()
corr_pairs = correlation_mat.unstack()

pd.set_option('display.max_rows', None)
print(corr_pairs)
```

```python
# 
```

```python
sorted_pairs = corr_pairs.sort_values()

pd.set_option('display.max_rows', None)
print(sorted_pairs)
```

```python
# Look for high correlations (<0.5)
strong_pairs = sorted_pairs[abs(sorted_pairs > 0.5)]  

pd.set_option('display.max_rows', None)
print(strong_pairs)
```

```python
# Looking at the top 15 companies
CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]
CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64')

print(CompanyGrossSumSorted)
```
